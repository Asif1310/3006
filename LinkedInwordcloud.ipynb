```python
import re # regex
import nltk
nltk.download('stopwords') # because we don't want our word clouds to be filled with "I", "AND", "OR", etc.
from nltk.corpus import stopwords 
corpus = []
all_stopwords = stopwords.words('english') + ['would', 'could', 'one', 'hence']
for i in range(0, len(df)-1):
  cont = re.sub('[^a-zA-Z]', ' ', df['TEXT'][i] ) # we are using regex to remove any special characters that are not alphabets
  cont = cont.lower() # we are converting the text to lower case
  corpus.append(cont)
```

```python
import matplotlib.pyplot as plt # Our go-to library for all the plots (and because I never learned Seaborn)
!pip install wordcloud 
from wordcloud import WordCloud
```

```python
wordcloud = WordCloud(background_color='white', stopwords = all_stopwords).generate(str(corpus))
```

```python
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show() # to display the result of 
```
